```python
import os
import pandas as pd
```

```python
def generate_work_record_file(root_path, total_dialogue_count):
    df = pd.DataFrame(columns=["Dialogue ID", "負責人", "語者一[性別]", "語者一[年紀]", "語者二[性別]", "語者二[年紀]", "對話生成", "情感標註"])
    dialogue_ids = [f"dialogue-{str(i).zfill(4)}" for i in range(0, total_dialogue_count)]
    df["Dialogue ID"] = dialogue_ids
    file_path = os.path.join(root_path, "對話生成工作紀錄.xlsx")
    # Construct ExcelWriter
    writer = pd.ExcelWriter(file_path)
    # Save DataFrame to Excel File
    df.to_excel(writer, sheet_name='Sheet1', encoding='utf-8', index=False)
    # Load the worksheet
    worksheet = writer.sheets['Sheet1']
    # Set column width
    worksheet.set_column('A:F', 17)
    worksheet.set_column('G:H', 10)
    writer.save()
    return df
```

```python
total_dialogue_count = 2000
root_path = "/content/drive/MyDrive/論文資料完整版"
df = generate_work_record_file(root_path, total_dialogue_count)
```
